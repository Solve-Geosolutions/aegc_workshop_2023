### Get the Lithology Name for Each Tile and Save It in a New Pickle File
This notebook will find the lithology name for tiles with available lithology label and saves them together with the tile information and SSL features. 

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

Load data

```python
# Load the spread sheets
df = pd.read_pickle(r'../outputs/features.pkl')
meta = pd.read_csv(r'../data/tiles_stats.csv')
chem = pd.read_csv(r"../data/filt_chem.csv")
```

```python
# Print the total number of rows in the spread sheet
print(df.shape[0])
# Show the first 5 lines in the spread sheet
df.head(5)
```

```python
# Show the first 5 lines in the spread sheet
meta.head(5)
```

```python
# Show the first 5 lines in the spread sheet
chem.head(5)
```

```python
# Merge the two spread sheets
meta_chem_joined = meta.merge(chem, on='SAMPLE_NO', how='inner')
meta_chem_joined = meta_chem_joined[['tile_filename', 'LITHOLOGY_NAME']]
print(meta_chem_joined.shape[0])

# Show the first 5 lines in the merged spread sheet
meta_chem_joined.head(5)
```

```python
# Merge the two spead sheets 
meta_chem_joined = meta_chem_joined.rename(columns={'tile_filename':'filename'})
df_meta_chem_joined = df.merge(meta_chem_joined, on='filename', how='inner')
df_meta_chem_joined = df_meta_chem_joined.rename(columns={'LITHOLOGY_NAME':'lithology_name'})
print(df_meta_chem_joined.shape[0])

# Show the first 5 lines in the merged spread sheet
df_meta_chem_joined.head(5)
```

```python
# Save the features together with their Lithology Label in a new pickle file
df_meta_chem_joined.to_pickle('../outputs/features_litho.pkl')
```
