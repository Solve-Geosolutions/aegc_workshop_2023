### Select Coherent Tiles
This notebook will select the coherent tiles and saves them in a new directory. The coherence of the tiles are given as an input to the code. The code will save tiles with more than 90% coherence in the new directory. 

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
```

Load data

```python
# Open the spread sheet
chem = pd.read_csv(r"../data/images_chem_samples.csv")
```

```python
# Show the first 5 rows in the spread sheet
chem.head(5)
```

```python
import shutil
import os

chem_coher = chem[chem['coherent_rock']>0.9]
source = '../data/processed/tiles'
target = '../data/processed/coherent_tiles'

# Create the output directory
os.makedirs(target, exist_ok=True)

# Save coherent tiles in the output directory
for index, row in chem_coher.iterrows():
    src = os.path.join(source, row['tile_filename'].split('_')[0], row['tile_filename'])
    dst = os.path.join(target, row['tile_filename'])
    shutil.copyfile(src, dst)
```
